```python
import numpy as np
from sktime.transformations.series.detrend import Detrender,Deseasonalizer
import matplotlib.pyplot as plt
import pandas as pd
```

### First we try with removing the deseasonality

```python
x = np.arange(0,100)
y = 4*x
plt.plot(x,y)
```

```python
y = pd.DataFrame(y)
transformer = Detrender()
y_hat = transformer.fit_transform(y)
#y_hat = transformer.fit_transform(y_hat)
plt.plot(x,y_hat)
plt.ylim(-1,1)
plt.show()
print(y_hat[0:5])
print(y[0:5])
```

```python
y_new = transformer.inverse_transform(y_hat)
plt.plot(x,y_new)
```

```python
x = np.arange(0,20,0.5)*np.pi/2
y = pd.Series(2*np.sin(x)+np.arange(0,20,0.5)+np.random.uniform(0,2,len(x)))
plt.plot(x,y)
```
